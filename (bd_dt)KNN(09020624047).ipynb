```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score
#mengimpor library seperti pandas untuk baca data, train_test_split untuk membagi data latih dan uji,
#KNeighborsClassifier sebagai algoritma klasifikasi berbasis jarak ke tetangga terdekat, StandardScaler untuk menormalkan skala fitur numerik,
#LabelEncoder untuk mengubah data kategorikal menjadi angka, serta accuracy_score untuk mengukur akurasi model.
```

```python
df = pd.read_excel('datasetknnfix(09020624047).xlsx')
print(df.head())
#membaca file membaca dan menyimpannya ke dalam variabel df (DataFrame), lalu menampilkan 5 baris pertama dari data tersebut menggunakan print(df.head()) untuk melihat sekilas isi dataset.
```

```python
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
df['smoking_history'] = le.fit_transform(df['smoking_history'])
X = df.drop('diabetes', axis=1)
y = df['diabetes']
#mengubah data kategorikal menjadi numerik menggunakan LabelEncoder pada kolom gender dan smoking_history (misal: Laki=0, Perempuan=1), lalu memisahkan fitur (X) dengan target (y).
#X berisi semua kolom kecuali diabetes yang dihapus menggunakan df.drop(), sedangkan y hanya berisi kolom diabetes sebagai label yang akan diprediksi.
```

```python
print("Fitur:", X.columns.tolist())
print("Jumlah data:", X.shape)
#menampilkan daftar nama-nama fitur (kolom) yang digunakan sebagai variabel X dalam bentuk list, serta menampilkan jumlah data (baris) dan jumlah fitur (kolom) dalam bentuk tuple (baris, kolom) menggunakan atribut shape.
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
#menormalisasi fitur numerik menggunakan StandardScaler agar semua fitur memiliki skala yang sama (rata-rata 0 dan standar deviasi 1), lalu menyimpan hasilnya ke X_scaled.
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
#train_test_split memisahkan x (fitur) dan y (target) secara acak. Parameter test_size=0.2 berarti 20% data akan menjadi data uji, sisanya 80% menjadi data latih.
#random_state=42 digunakan agar hasil pembagian selalu sama setiap kali kode dijalankan
```

```python
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
#membuat model KNN dengan jumlah tetangga terdekat (K) sebanyak 5, lalu melatih model tersebut menggunakan data latih (X_train sebagai fitur dan y_train sebagai target).
#Proses fit() akan membuat model mempelajari pola dari data latih dengan menyimpan seluruh data latih untuk nantinya digunakan saat prediksi.
```

```python
y_pred = model.predict(X_test)
print("Akurasi:", accuracy_score(y_test, y_pred))
#Model memprediksi data uji (X_test), lalu akurasi dihitung dengan membandingkan hasil prediksi (y_pred) terhadap label asli (y_test) dan ditampilkan.
#hasilnya 96.135%, yang berarti dari seluruh data uji, model berhasil memprediksi dengan benar sebanyak 96 dari 100 kasus. Hasil ini tergolong sangat tinggi karena beberapa faktor — fitur-fitur yang digunakan seperti HbA1c_level, blood_glucose_level, dan bmi memiliki korelasi yang kuat terhadap kondisi diabetes, sehingga KNN mudah menemukan pola kedekatan antar data.
```
